```python
# Data Visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  1. Line Chart  Trend of Revenue Over Time

df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], dayfirst=True)
df_by_date = df_cleaned.groupby('Date')['Revenue'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=df_by_date, x='Date', y='Revenue', color='teal')
plt.title("Revenue Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Revenue")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
 # 2. Bar Chart  Average Units Sold per Product Category

avg_units = df_cleaned.groupby('Category')['Units Sold'].mean().sort_values()

plt.figure(figsize=(10,6))
avg_units.plot(kind='barh', color='skyblue')
plt.title("Average Units Sold per Category")
plt.xlabel("Average Units Sold")
plt.ylabel("Product Category")
plt.tight_layout()
plt.show()
```

```python
#  3. Histogram  Distribution of Discounts
plt.figure(figsize=(8,5))
sns.histplot(df_cleaned['Discount'], bins=20, kde=True, color='orange')
plt.title("Distribution of Discounts")
plt.xlabel("Discount")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
#  4. Scatter Plot  Price vs Revenue

plt.figure(figsize=(8,6))
sns.scatterplot(data=df_cleaned, x='Price', y='Revenue', hue='Category', palette='Set2')
plt.title("Price vs Revenue by Category")
plt.xlabel("Price")
plt.ylabel("Revenue")
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
